```python
import numpy as np
# import sys
# sys.path.append("../")
from matviz.histogram_utils import nhist
```

```python
A = {'mu_is_Zero': np.random.randn(10 ** 5), 'mu_is_Two': np.random.randn(10 ** 3) + 2}
```

```python
_ = nhist(A)
```

```python
# change the colormap
_ = nhist(A, color='viridis')
```

```python
# about 4 times the recommended number of bins
_ = nhist(A, f=4)
```

```python
# all histograms using identical bins
_ = nhist(A, same_bins_flag=True)
```

```python
# remove the default error bar from the top
_ = nhist(A, noerror=True)
```
